```python
import matplotlib.pyplot as plt

%matplotlib inline

import numpy as  np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
##plotting using matplotlib

plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
y=x*x
plt.plot(x,y,'ro--')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,a,'g')
plt.subplot(2,2,3)
plt.plot(y,b,'y')
```

```python
#Compute the x and y coordinates for points on a sine curve
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
```
